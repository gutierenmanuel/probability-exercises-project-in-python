## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np

cantidad_repeticiones = 10000

mayor_que_7 = 0
numero_par = 0

mayor_y_par = 0

for i in range(cantidad_repeticiones):
    dado1 = np.random.randint(1,7)
    dado2 = np.random.randint(1,7)

    suma = dado1 + dado2


    if suma > 7 :
        mayor_que_7 += 1

    if suma % 2 == 0:
        numero_par += 1

    if suma % 2 == 0 or suma > 7:
        mayor_y_par += 1


print(f'el numero de veces que los dados han sumado mas de 7 es {mayor_que_7}')
print(f'la cantidad relativa de veces es {mayor_que_7/cantidad_repeticiones}')
print('\n')
print(f'el numero de veces que los dados han dado un numero par es {numero_par}')
print(f'la cantidad relativa de veces es {numero_par/cantidad_repeticiones}')
print('\n')
print(f'el numero de veces que los dados han dado un numero par y mas que 7 es {mayor_y_par}')
print(f'la cantidad relativa de veces es {mayor_y_par/cantidad_repeticiones}')
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
print('\n')

#Codigo:

#Inicializo variables

cantidad_repeticiones = 100000

tres_blancas_dos_rojas = 0
bolas_iguales = 0

#Creo una funcion para comprobar si son todos iguales
def todos_iguales(lista):
    primer_elemento = lista[0]
    for elemento in lista:
        if elemento != primer_elemento:
            return False
    return True


for i in range(cantidad_repeticiones):

    #inicializo las variables que se repetiran en el bucle
    bolas_blancas = 0
    bolas_rojas = 0

    lista_bolas = []

    #Selecciono las bolas al azar
    bolas_escogidas = np.random.randint(0,60,5)

    #Guardo en una lista el nombre de las bolas
    for bola in bolas_escogidas:
        lista_bolas.append(ball_box[bola])

    #Compruebo las bolas blancas y rojas
    for bola in lista_bolas:
        if bola == 'White':
            bolas_blancas += 1
        elif bola == 'Red':
            bolas_rojas += 1

    #Guardo el resultado
    if bolas_blancas == 3 and bolas_rojas == 2:
        tres_blancas_dos_rojas +=1

    #Compruebo que las bolas sean iguales y guardo el resultado
    if todos_iguales(lista_bolas):
        bolas_iguales +=1

#Imprimo los resultados

print(f'el numero de veces que salen 3 blancas y 2 rojas es {tres_blancas_dos_rojas}')
print(f'la cantidad relativa de veces es {tres_blancas_dos_rojas/cantidad_repeticiones}')
print('\n')
print(f'el numero de veces en el que todas las bolas son iguales es {bolas_iguales}')
print(f'la cantidad relativa de veces es {bolas_iguales/cantidad_repeticiones}')
print('\n')

```
